```python
import pandas as pd

deg_sig3 = pd.read_csv("deg_sig3.csv", index_col="Unnamed: 0")
deg_sig_vst = pd.read_csv("deg_sig_vst.csv", index_col="Unnamed: 0")
integ_final_survival_info = pd.read_csv("integ_final_survival_info.csv", index_col="Unnamed: 0")
```

```python
sorted_deg_sig3 = deg_sig3.sort_values("padj", ascending=True)
top20_padj = sorted_deg_sig3[0:20]
top20_padj
```

```python
top20_padj.loc[:, ["symbol", "padj"]]
```

```python
a1 = deg_sig_vst.loc[top20_padj.index, :]
a1.index = top20_padj['symbol']
a1
```

```python
transpose_a1 = a1.transpose()
new_a1 = transpose_a1.loc[integ_final_survival_info.index, :]
```

```python
final_df = pd.concat([new_a1, integ_final_survival_info['integ_cluster']], axis=1)
final_df.to_csv("final_df.csv", index=True)
pd.read_csv("final_df.csv", index_col="Unnamed: 0")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.metrics import classification_report

X = final_df.loc[:, final_df.columns != "integ_cluster"]
y = final_df['integ_cluster'].astype('category')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
cl = RandomForestClassifier()
param_grids = {
    "class_weight": ['balanced', 'balanced_subsample', None],
    "n_estimators": [100, 300, 500],
    "max_depth": [None, 3, 5]
}
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

gridsearch = GridSearchCV(cl, param_grids, cv=kfold, n_jobs=-1)
gridsearch.fit(X_train, y_train)

print("Best parameters:", gridsearch.best_params_)
print("Best score:", gridsearch.best_score_)
```

```python
import optuna
from sklearn.model_selection import cross_val_score

def objective(trial):
    rf_max_depth = trial.suggest_int("rf_max_depth", 2, 32)
    rf_n_estimators = trial.suggest_int("rf_n_estimators", 10, 200, log=True)
    rf_min_samples_split = trial.suggest_int("rf_min_samples_split", 2, 10)
    rf_min_samples_leaf = trial.suggest_int("rf_min_samples_leaf", 1, 10)
    
    model = RandomForestClassifier(
        max_depth = rf_max_depth,
        n_estimators = rf_n_estimators,
        min_samples_split = rf_min_samples_split,
        min_samples_leaf = rf_min_samples_leaf,
        random_state = 42
    )
    score = cross_val_score(model, X_train, y_train, n_jobs=-1, cv=3)
    accuracy = score.mean()
    
    return accuracy
```

```python
study = optuna.create_study(direction="maximize")
optuna.logging.set_verbosity(optuna.logging.WARNING)
study.optimize(objective, n_trials=100, show_progress_bar=True)
print(study.best_trial) #0.661
```

```python
from sklearn.metrics import accuracy_score

rfc = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    class_weight=None,
    n_jobs=-1,
    random_state=42
)

rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(accuracy_score(y_test, y_pred))

# rfc.fit(X_train.loc[:, ['EPAS1', 'B3GNT3', 'TOP2A', 'PECAM1', 'TEDC2', 'EMP2', 'S1PR1', 'FAM83A', 'AFAP1-AS1', 'PYCR1', 'RGCC', 'ETV4', 'STX11','FAM83A-AS1', 'IQGAP3']], y_train)
# y_pred2 = rfc.predict(X_test.loc[:, ['EPAS1', 'B3GNT3', 'TOP2A', 'PECAM1', 'TEDC2', 'EMP2', 'S1PR1', 'FAM83A', 'AFAP1-AS1', 'PYCR1', 'RGCC', 'ETV4', 'STX11','FAM83A-AS1', 'IQGAP3']])
# print(accuracy_score(y_test, y_pred2))

importances = rfc.feature_importances_
feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False)
#print(sum(feat_imp[:10])) #0.605
#print(feat_imp[0:15].index)
```

```python
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt

top10_feat_imp = feat_imp[:10].index
cluster_group = X.groupby(y)[top10_feat_imp].mean()
cluster_group_z = cluster_group.apply(zscore, axis=0)

sns.heatmap(
    cluster_group_z,
    annot = True,
    fmt = ".2f",
    cmap = "RdBu_r",
    center = 0,
    cbar_kws={"label": "z-score"}
)

plt.xlabel('Top10_feature_importance_genes')
plt.ylabel('Clusters')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss_top10 = pd.DataFrame(
    ss.fit_transform(final_df[list(top10_feat_imp)]),
    index=final_df[list(top10_feat_imp)].index,
    columns=top10_feat_imp
)
```

```python
integ_final_survival_info = pd.read_csv('integ_final_survival_info.csv', index_col="Unnamed: 0")
df_top10 = pd.concat([ss_top10, integ_final_survival_info[['time', 'event', 'integ_cluster']]], axis=1)
```

```python
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
import numpy as np

km = KaplanMeierFitter()

for i in df_top10['integ_cluster'].unique():
    mask = df_top10['integ_cluster'] == i
    km.fit(
        df_top10.loc[mask, 'time'],
        df_top10.loc[mask, 'event'],
        label=f'cluster {i}'
    )
    km.plot_survival_function()
    
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.xticks(np.arange(0, 7000, 365))
plt.title('KM from RF selected genes')
fig = plt.gcf()
fig.set_size_inches(20, 10)
plt.show()
```

```python
import lifelines
from lifelines import CoxPHFitter

cox_df = df_top10[['time', 'event'] + top10_feat_imp.tolist()]
cp = CoxPHFitter()
cp.fit(cox_df, duration_col='time', event_col='event') 
cp.print_summary()
```

```python
cp.summary[['exp(coef)', 'p']].sort_values('exp(coef)', ascending=False).to_csv('cp_summary.csv', index='covariate')
cp_summary = pd.read_csv('cp_summary.csv', index_col='covariate')
```
